```python
def recommender_function(df,movie_name):
  selected_features = ["genres", "keywords", "tagline", "cast", "director"]

  for feature in selected_features:
    df[feature] = df[feature].fillna('')

  combined_features = df["genres"] + " " +df["keywords"]+ " " +df["tagline"]+ " " +df["cast"]+ " " +df["director"]

  vectorizer = TfidfVectorizer(stop_words="english")

  features_vectors = vectorizer.fit_transform(combined_features)

  cosine_sim = cosine_similarity(features_vectors,
                               features_vectors)

  listofalltitles = df["title"].tolist()

  find_the_close_match = difflib.get_close_matches(movie_name,listofalltitles)

  close_match = find_the_close_match[0]

  index_of_the_movie = df[df.title == close_match]["index"].values[0]

  similarity_score = list(enumerate(cosine_sim[index_of_the_movie]))

  sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)

  second_values = [item[0] for item in sorted_similar_movies[1:]]

  recommended_movies = df[df["index"].isin(second_values[0:10])]["title"]

  return recommended_movies
```

```python
recommender_function(df,"iron man")
```
